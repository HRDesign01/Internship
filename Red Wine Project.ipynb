With this, we are able to give statistical representation of the dataset.Based on the outcome we can say the dataset looks perfect because no negative/invalid value is present.
From the above, we can also observe that:
1. The count row has the same values which means there are no missing values present.
2. The mean value is greater than the median in the fixed acidity, volatile acidity, citric acid, residual sugar, chlorides, free sulfur dioxide, total sulfur dioxide, pH, sulphates, and alcohol column. This mean that the data is skewed to the right.
3. The mean value is less than median in the density and quality column. this means that the data is skew to the left.
4. After further look into the dataset, we can observe there are significant differences between the 75% and max value. With this observation, we can say there are outliers present in the dataset.
5. We are also able to see the values of the standard deviation, min, and 25% percentile values base on this describe method.
